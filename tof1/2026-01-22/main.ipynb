```python
import pandas as pd
import matplotlib.pyplot as plt

# leser filen med pandas, endret på filen slik at den har en header
df = pd.read_csv("Timersov.txt")
df
```

```python
df.plot()

# xticks til hele tall fra minste verdien i datasettet til høyeste pluss 1 for litt marginer
plt.xlabel("Dager")
plt.xticks(range(int(df.index.min()), int(df.index.max()) + 1))
plt.ylabel("Timer søvn")

# finner gjennomsnitt med bruk av .mean() og ekstraherer ut pandas listen sin første verdi med .iloc[0]
avg = df.mean().iloc[0]

# plotter gjennomsnittet som en dotted line
plt.axhline(avg, linestyle="--", linewidth=2, color="gray", label=f"Gjennomsnitt: {avg:.2f} timer søvn")

plt.legend()
plt.show()
```

Valgte et linje diagram da det beskriver søvn over tid i en periode.
Kan også velge et søyle diagram som kan være bedre da det mer korrekt viser at det er per dag. Da ser det slik ut:

```python
df.plot(kind="bar")
plt.xlabel("Dager")
plt.xticks(range(int(df.index.min()), int(df.index.max()) + 1))
plt.ylabel("Timer søvn")

avg = df.mean().iloc[0]
plt.axhline(avg, linestyle="--", linewidth=2, color="gray", label=f"Gjennomsnitt: {avg:.2f} timer søvn")

plt.legend()
plt.show()
```

Kan også lage en fil `Timersov-2.txt` hvor de to høyeste verdiene er fjernet da 24 timer og 36 timer søvn er ekstremt mye. Da synker snittet fra 9 timer søvn til 7 timer søvn med de to største verdiene fjernet. Valgte å ikke fjerne de minste verdiene da de minste verdiene er ikke like urealistiske, synes jeg.
