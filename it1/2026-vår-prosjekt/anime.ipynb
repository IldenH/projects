```python
from fastai.collab import *
from fastai.tabular.all import *
```

```python
import json
import pandas as pd

meta = None
rows = []

with open("anime-offline-database.jsonl", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if not line.strip():
            continue
        
        obj = json.loads(line)
        
        if i == 0:
            meta = obj
        else:
            rows.append(obj)

df = pd.DataFrame(rows)

print(meta)
df.head()
```

```python
user_df = pd.read_csv("users-score-2023.csv")
```

```python
user_df.head()
```

```python
import numpy as np

df = user_df.copy()

df = df[['user_id', 'Anime Title', 'rating']].copy()
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna(subset=['user_id', 'Anime Title', 'rating'])

df['user_id'] = df['user_id'].astype(str)      # often safer for categorical IDs
df['Anime Title'] = df['Anime Title'].astype(str)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df = df.dropna(subset=['rating'])

print(df.isna().sum())
print(df['rating'].min(), df['rating'].max())
```

```python
dls = CollabDataLoaders.from_df(
    df,
    user_name="user_id",
    item_name="Anime Title",
    rating_name="rating",
    bs=64
)
dls.show_batch()
```

```python
user_df["rating"].value_counts()
```

```python
learn = collab_learner(dls, n_factors=50, y_range=(1, 10.5))
learn.fit_one_cycle(5, 5e-3, wd=0.1)
```

```python
learn.model
```

```python
g = df.groupby('Anime Title')['rating'].count()
top_movies = g.sort_values(ascending=False).index.values[:1000]
top_idxs = tensor([learn.dls.classes['Anime Title'].o2i[m] for m in top_movies])
movie_w = learn.model.i_weight.weight[top_idxs].cpu().detach()
movie_pca = movie_w.pca(3)
fac0,fac1,fac2 = movie_pca.t()
idxs = list(range(50))
X = fac0[idxs]
Y = fac2[idxs]
plt.figure(figsize=(12,12))
plt.scatter(X, Y)
for i, x, y in zip(top_movies[idxs], X, Y):
    plt.text(x,y,i, color=np.random.rand(3)*0.7, fontsize=11)
plt.show()
```

```python
learn.show_results()
```

```python
learn.model.eval()
```

```python
item_emb = learn.model.i_weight.weight.detach().cpu().numpy()
item_bias = learn.model.i_bias.weight.detach().cpu().numpy().squeeze()

anime_vocab = dls.classes['Anime Title']

anime_title_to_index = {str(title): int(i) for i, title in enumerate(anime_vocab)}
anime_index_to_title = {str(i): str(title) for i, title in enumerate(anime_vocab)}

path = Path("export")
path.mkdir(exist_ok=True)

np.save(path/"item_embeddings.npy", item_emb)
np.save(path/"item_bias.npy", item_bias)

with open(path/"anime_title_to_index.json", "w", encoding="utf-8") as f:
    json.dump(anime_title_to_index, f, ensure_ascii=False)

with open(path/"anime_index_to_title.json", "w", encoding="utf-8") as f:
    json.dump(anime_index_to_title, f, ensure_ascii=False)
```
