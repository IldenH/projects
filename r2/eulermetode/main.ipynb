```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("vanntemperatur-kaldtvann.csv", delimiter=";")
df
```

```python
df["Temperatur"].plot()
```

```python
tid = df["Tid"].tolist()
temperatur = df["Temperatur"].tolist()

dx = 0.1

k = -0.008
T = 27 # romtemp

xs = []
ys = []

x = tid[0]
y = temperatur[0]

while x <= tid[-1]:
    xs.append(x)
    ys.append(y)

    x += dx
    y += dx * k * (y - T)

plt.plot(tid, temperatur, 'x')
plt.plot(xs,ys)
plt.show()
```
