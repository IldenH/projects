```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
T = 3  # sekunder
t = np.linspace(0, T, n)
dt = T / n

s = np.zeros(n)
v = np.zeros(n)
a = np.zeros(n)

m = 0.20  # kg
g = 9.81  # m/s^2
k = 0.01  # N/(m/s)^2
v[0] = 10 # m/s

for i in range(n - 1):
    G = g * m
    L = k * (v[i] ** 2) # L = kv²
    F_sum = G - L
    a[i] = -F_sum / m
    v[i + 1] = a[i] * dt + v[i] # v = at + v_0
    s[i + 1] = v[i + 1] * dt + s[i] # s = vt + s_0
```

```python
plt.grid()
plt.title("Posisjon")
plt.xlabel("$t$ / s")
plt.ylabel("$s$ / m")
plt.plot(t, s, label="k = %.2f" % k)
plt.legend()
```

```python
plt.grid()
plt.title("Fart")
plt.xlabel("$t$ / s")
plt.ylabel("$v$ / (m/s)")
plt.plot(t, v, label="k = %.2f" % k)
plt.legend()
```

```python
plt.figure(3)
plt.grid()
plt.title("Akselerasjon")
plt.xlabel("$t$ / s")
plt.ylabel("$a$ / (m/s$^2$)")
plt.plot(t[:-1], a[:-1], label="k = %.2f" % k)
plt.legend()
```
